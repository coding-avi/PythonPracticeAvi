### please make a simple calculator program
###### it will ask you to enter number 1 and then ask to enter number 2
###### print the sum, diff, product and division

```python
x = int(input("please enter first number: "))
y = int(input("please enter second number: "))

print(f"sum of {x} and {y} is {x+y}")
print(f"difference of {x} and {y} is {x-y}")
print(f"product of {x} and {y} is {x*y}")
```
