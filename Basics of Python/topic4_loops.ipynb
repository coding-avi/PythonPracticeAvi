```python
import random
```

```python
name = input("name:? ")
number = int(input("please type a number one through five "))
options = [1,2,3,4,5]
computer = random.choice(options)

print(f"{name} chooses {number}")
print(f"computer chooses {computer}")
```

```python
game = input("game:? ")


for i in range(3):
    human = input("choose rock paper or scissors")
    options = ["rock", "paper", "scissors"]
    computer = random.choice(options)

    print(f"{game} chooses {human}")
    print(f"computer choose {computer}")


    if human == computer:
        print("draw")
    elif human == "paper" and computer =="rock" or human =="rock" and computer=="scissors" or human=="scissor" and computer =="paper":
        print(f"{game} won")

    else:
        print("computer won!")
```

```python
# guesssing game poweful

computer = random.randint(1,100)
name = input("name:? ")

for i in range(10):
    
    number = int(input("please type a number one through hundred "))
    if number < computer:
        print('increase your number!')
    elif number > computer:
        print("decrease your number!")
    else:
        print("matched!!!")
        break

print("actually number was",computer)
print("you guess it in ",i,"trials")
```
