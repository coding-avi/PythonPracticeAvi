```python
class Cat:
    def __init__(self,name,status):
        self.name = name
        self.status = status
    def tell(self):
        print(f"{self.name} is now {self.status} ")
```

```python
c1 = Cat("lily","eating")
c2 = Cat("don","eating")
c3 = Cat("cato","flying")
```

```python
c2.tell()
```

```python
c1.tell()
```

```python
class Animal:
    def __init__(self,name,legs):
        self.name = name
        self.legs = legs  
    def info(self):
        print(f"{self.name} has {self.legs} legs")
```

```python
a1 = Animal("Tiger",4)
a2 = Animal("lizard",5)
a3 = Animal("chicken",2)
```
