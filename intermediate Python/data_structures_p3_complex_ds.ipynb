```python
# we make a program in which if uder enter somethign it will keep on adding it inn list until user enters "stop"
```

```python
x = input("Enter something (type 'stop' to end): ")
my_list = []
while x.lower() != "stop":
    my_list.append(x)
    x = input("Enter something (type 'stop' to end): ")
print("You entered:", my_list)
print("Maximum value:", max(my_list))
```
