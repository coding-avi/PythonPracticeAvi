```python
animals = ["cat", "dog", "rabbit", "hamster", "parrot", "fish"]
print(animals[5])  # Output: fish
```

```python
animals[5] = "penguin"
print(animals)  # Output: ['cat', 'dog', 'rabbit', 'hamster', 'parrot', 'penguin']
```

```python
animals[2] = "chicken"
print(animals)  # Output: ['cat', 'dog', 'chicken', 'hamster', 'parrot', 'penguin']
```

```python
# lets remove now penguin
del animals[5]
print(animals)  # Output: ['cat', 'dog', 'chicken', 'hamster', 'parrot']
```

```python
del animals[2]  
print(animals)  # Output: ['cat', 'dog', 'hamster', 'parrot']
```
